```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def shape_space(psi, i, n, A):
    return A[i] * math.comb(n, i) * psi**i * (1-psi)**(n-i)

def class_func(psi):
    return math.sqrt(psi) * (1-psi)

def thickness(psi, n, A):
    return sum([shape_space(psi, i, n, A) * class_func(psi) for i in range(n)])
```

```python
N = 3
NUM_POINTS = 1000

rng = np.random.default_rng()

As = rng.uniform(-1, 1, size=N+1)
Ap = rng.uniform(-1, 1, size=N+1)

# ensure continuous curvature at leading and trailing edge
Ap[0] = As[0]
Ap[-1] = As[-1]

x = np.linspace(0, 1, NUM_POINTS)
ss = [thickness(v, 2, As) for v in x]
ps = [-thickness(v, 2, Ap) for v in x]
```

```python
plt.axis('equal')
plt.plot(x, ss, x, ps)
```
