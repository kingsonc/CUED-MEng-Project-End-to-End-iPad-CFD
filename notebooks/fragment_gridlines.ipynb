```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fract(x):
    return x - np.floor(x)

def clamp(x, minVal, maxVal):
    return min(max(x, minVal), maxVal)

def smoothstep(edge0, edge1, x):
    t = clamp((x - edge0) / (edge1 - edge0), 0.0, 1.0)
    return t * t * (3.0 - 2.0 * t)
```

```python
fr = []
fr_abs = []
line = []

for i in np.arange(0, 5, 0.01):
    fr_val = fract(i) - 0.5
    fr_abs_val = np.abs(fr_val)
    line_val = smoothstep(2, 6, 8*fr_abs_val)

    fr.append(fr_val)
    fr_abs.append(fr_abs_val)
    line.append(line_val)
```

```python
plt.rcParams.update({'font.size': 20})

fig, axs = plt.subplots(3, figsize=(16,10), sharex=True)

axs[0].plot(x, line)
axs[1].plot(x, fr_abs)
axs[2].plot(x, fr)

axs[0].set(ylabel="smoothstep")
axs[1].set(ylabel="abs fract")
axs[2].set(xlabel="x", ylabel="fract")

fig.align_ylabels(axs)
```
